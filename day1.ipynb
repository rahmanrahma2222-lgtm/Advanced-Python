```python
numbers=['9876543210','12453','9988776655','9876501234']
count=0
valid_count=0
invalid_count=0
valid_nums=set()
for nm in numbers:
  if len(nm)!=10:
   invalid_count+=1
   continue
  else:
   print(nm,'valid number')
   valid_count+=1
  valid_nums.add(nm)

print(f'count of invalid is',invalid_count)
print(f'count of valid is',valid_count)
print(valid_nums)
```
